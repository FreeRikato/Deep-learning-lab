```python
import numpy as np
x = []
y = []
for i in range(0, 100, 5):
  a = []
  for j in range(i, i + 20, 5):
    a.append(j)
  x.append(a)
x = np.array(x)
for i in range(20, 116, 5):
  y.append(i)
y = np.array(y)
```

```python
x = x.reshape(20, 4, 1)
```

```python
x
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model=Sequential()
model.add(LSTM(50,activation='relu',input_shape=(4,1)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=300,verbose=1)
```

```python
test=np.array([85,90,95,100])
test=test.reshape(1,4,1)
res = model.predict(test)
res
```
